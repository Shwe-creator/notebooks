```python
import cPickle as pickle
import numpy as np

with open('./timit_embeddings.pkl', 'rb') as f:
    embeddings = pickle.load(f)
```

```python
metadata = [','.join(x[0]) for x in embeddings]
words = [x[1] for x in embeddings]
features_c = np.array([x[2].flatten() for x in embeddings])
features_h = np.array([x[3].flatten() for x in embeddings])
```

```python
from sklearn.manifold import TSNE
from bokeh.plotting import output_notebook, figure, show
from bokeh.models import HoverTool, TapTool, ColumnDataSource, OpenURL
from bokeh.charts import Scatter

output_notebook()
```

```python
tsne = TSNE(n_components=2)
data_reduced_h_top = tsne.fit_transform(features_h[:, 200:300])

fig = figure(plot_width=800, plot_height=600, title=None, tools=['hover', 'box_zoom', 'reset'])
fig.scatter(x=data_reduced_h_top[:, 0], y=data_reduced_h_top[:, 1], source=ColumnDataSource(data={"word": words, "metadata": metadata}))
fig.select(HoverTool).tooltips = [('word', '@word'), ('metadata', '@metadata')]
show(fig)
```

```python
tsne = TSNE(n_components=2)
data_reduced_c_top = tsne.fit_transform(features_c[:, 200:300])

fig = figure(plot_width=800, plot_height=600, title=None, tools=['hover', 'box_zoom', 'reset'])
fig.scatter(x=data_reduced_c_top[:, 0], y=data_reduced_c_top[:, 1], source=ColumnDataSource(data={"word": words, "metadata": metadata}))
fig.select(HoverTool).tooltips = [('word', '@word'), ('metadata', '@metadata')]
show(fig)
```
