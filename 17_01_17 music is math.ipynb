```python
nmeasures_to_successes = {}
for nmeasures in range(1, 13):
    subdivs_total = nmeasures * 16
    successes = []
    for seqlen in range(1, 13):
        for nrepeats in range(1, 3):
            seq_total = 0
            seq_current = seqlen
            niter = 0
            while seq_total < subdivs_total:
                seq_total += seq_current * nrepeats
                seq_current += 1
                niter += 1
            if seq_total == subdivs_total and niter > 1:
                successes.append((seqlen, nrepeats))
    nmeasures_to_successes[nmeasures] = successes
```

```python
def accumu(lis):
    total = 0
    for x in lis:
        total += x
        yield total
```

```python
from midiutil.MidiFile import MIDIFile

def sequence_to_midi(sequence, output_name):
    # create your MIDI object
    mf = MIDIFile(1)     # only 1 track
    track = 0   # the only track

    time = 0    # start at the beginning
    mf.addTrackName(track, time, "mathmusic")
    mf.addTempo(track, time, 160)

    channel = 0
    volume = 127
    time = 0
    duration = 0.25
    for note in sequence:
        if note > 0:
            mf.addNote(track, channel, note, time, duration, volume)
        time += duration

    # write it to disk
    with open(output_name, 'wb') as outf:
        mf.writeFile(outf)
```
