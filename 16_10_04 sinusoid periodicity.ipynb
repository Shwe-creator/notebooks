```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 9)
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 2.0
f2 = 7.0
fs = 1024.0

dt = 1.0 / fs
t = np.arange(0.0, 2.0, dt)

x1 = np.sin(2.0 * np.pi * f1 * t)
x2 = np.sin(2.0 * np.pi * f2 * t)
plt.figure()
plt.plot(x1)
plt.figure()
plt.plot(x2)

x3 = x1 + x2
plt.figure()
plt.plot(x3)
```

```python
f = np.array([1.0, 4.0, 7.0, 11.0, 17.0])[np.newaxis]
fs = 1000.0
dt = 1.0 / fs
tmax = 2.0

t = np.arange(0.0, tmax, dt)[np.newaxis]

xarr = np.sin(2.0 * np.pi * f.T * t)
x = np.sum(xarr, axis=0)
plt.figure()
for sig in xarr:
    plt.plot(sig)
plt.plot(x)
```

The period of a signal is 1/

```python
X = np.fft.fft(x[:int(fs)])
Xmag = np.absolute(X)
Xphs = np.angle(X)
plt.figure()
plt.plot(Xmag)
plt.xlim([0, 20])
plt.figure()
plt.plot(Xphs)
```
