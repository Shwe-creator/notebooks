```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 9)
```

```python
f1 = 2.0
f2 = 7.0
fs = 1024.0

dt = 1.0 / fs
t = np.arange(0.0, 2.0, dt)

x1 = np.sin(2.0 * np.pi * f1 * t)
x2 = np.sin(2.0 * np.pi * f2 * t)
plt.figure()
plt.plot(x1)
plt.figure()
plt.plot(x2)

x3 = x1 + x2
plt.figure()
plt.plot(x3)
```

```python
f = np.array([1.0, 4.0, 7.0, 11.0, 17.0])[np.newaxis]
fs = 1000.0
dt = 1.0 / fs
tmax = 2.0

t = np.arange(0.0, tmax, dt)[np.newaxis]

xarr = np.sin(2.0 * np.pi * f.T * t)
x = np.sum(xarr, axis=0)
plt.figure()
for sig in xarr:
    plt.plot(sig)
plt.plot(x)
```

The period of a signal is 1/f0 where f0 is the fundamental. Integer numbers in Hz including prime numbers have a period of 1 second at the maximum

```python
X = np.fft.fft(x[:int(fs)])
Xmag = np.absolute(X)
Xphs = np.angle(X)
plt.figure()
plt.plot(Xmag)
plt.xlim([0, 20])
plt.figure()
plt.plot(Xphs)
```

```python
fs = 1000
f0 = 10.0
f1 = 30.0
dt = 1.0 / fs

t0 = np.arange(0.0, 1.0, dt)
t1 = np.arange(0.0, 2.0, dt)

x0 = np.sin(2.0 * np.pi * f0 * t0)
x1 = np.sin(2.0 * np.pi * f1 * t1)
x = np.zeros(len(x1))
x[:len(t0)] += x0
x += x1

plt.plot(x)

plt.figure()
plt.plot(np.absolute(np.fft.fft(x)))
plt.xlim([0, 70])
```

```python
FSQR = np.array([1.0, 3.0, 5.0, 7.0, 9.0, 11.0, 13.0, 15.0, 17.0, 19.0, 21.0])
FSAW = np.arange(1.0, 20.0, 1.0)

f0 = 10.0
f = FSAW[np.newaxis]
fs = 44100.0
tmax = 1.0

dt = 1.0 / fs
t = np.arange(0.0, tmax, dt)[np.newaxis]

a = np.eye(f.shape[1]) * 1.0 / (f)
xarr = np.dot(a, np.sin(2.0 * np.pi * f0 * np.dot(f.T, t)))
x = np.sum(xarr, axis=0)
plt.plot(x)
```
