```python
fs = 44100.0
dt = 1.0 / fs
periods = np.array([0.004, 0.006])
f = 1.0 / periods

t = np.arange(0.0, 2.0, dt)
xarr = np.sin(2.0 * np.pi * t[np.newaxis].T * f[np.newaxis])
x = np.sum(xarr, axis=1)
```

```python
%pylab inline

plt.plot(t, xarr[:, 0])
plt.plot(t, xarr[:, 1])
plt.plot(t, x)
plt.xlim([0.0, 0.02])
```
