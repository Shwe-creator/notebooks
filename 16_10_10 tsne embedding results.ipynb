```python
import cPickle as pickle
import numpy as np

with open('./embeddings.pkl', 'rb') as f:
    embeddings = pickle.load(f)

labels = [x[0] for x in embeddings]
data = np.array([x[1] for x in embeddings])
print data.shape
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
data_reduced = tsne.fit_transform(data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_reduced_pca = pca.fit_transform(data)
```

```python
from bokeh.plotting import output_notebook, figure, show
from bokeh.models import HoverTool, TapTool, ColumnDataSource, OpenURL
from bokeh.charts import Scatter

output_notebook()

fig = figure(plot_width=800, plot_height=600, title=None, tools=['hover', 'tap', 'box_zoom', 'reset'])
fig.scatter(x=data_reduced_pca[:, 0], y=data_reduced_pca[:, 1], source=ColumnDataSource(data={"name": labels}))
fig.select(HoverTool).tooltips = [('name', '@name')]
#fig.select(TapTool).callback = tap_callback
url = "http://rfs10k.ucsd.edu/files/rfs1k/wav/@name.wav"
fig.select(type=TapTool).callback = OpenURL(url=url)
show(fig)
```
