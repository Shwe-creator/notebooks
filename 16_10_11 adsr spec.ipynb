```python
from nputil import linterp
env = linterp(0.0, [(0.1, 1.0), (0.1, 0.75), (0.5, 0.75), (0.3, 0.0)], 1024)
plt.figure()
plt.plot(env)

ENV = np.fft.rfft(env)
ENV_M = np.absolute(ENV)
plt.figure()
plt.plot(ENV_M)
plt.xlim([0, 10])
```

```python
t = np.arange(0.0, 1.0, 1.0 / 1024)
x = np.sin(2.0 * np.pi * t * 30)
xenv = x * env
plt.figure()
plt.plot(xenv)

XENV = np.fft.rfft(xenv)
XENV_M = np.absolute(XENV)
plt.figure()
plt.plot(XENV_M)
plt.xlim([0, 45])
```
