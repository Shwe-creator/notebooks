```python
n = 1024
k = 4

x = np.identity(n)
```

```python
from itertools import cycle

matplotlib.rcParams.update({'font.size': 22})

def plot_xform(sk, xlim=None, ylim=(-1.1, 1.1)):
    lines = ["-","--","-.",":"]
    linecycler = cycle(lines)
    plt.figure(figsize=(12, 9))
    for i, s in enumerate(sk):
        plt.plot(np.real(s), linewidth=8, linestyle=next(linecycler), label=i)
    if xlim == None:
        xlim = (0, sk.shape[1])
    plt.xlim(xlim)
    plt.xticks([])
    plt.legend(loc='lower right')
    plt.ylim(ylim)
```

```python
s = np.fft.fft(x)
sk = s[:k]

plot_xform(sk)
```

```python
from scipy.fftpack import dct

s = dct(x)
plot_xform(s[:k], ylim=(-2.1, 2.1))
```
