```python
%pylab inline

import math
def calc_bmin(n, l, m):
    n, l, m = float(n), float(l), float(m)
    one = math.floor((l * n) / m)
    two = m / l
    return int(math.floor(one * two))

cascades = []
directs = []
for n in xrange(100):
    cascade = (n - calc_bmin(n, 1, 2)) + (n - calc_bmin(n, 1, 5)) + (n - calc_bmin(n, 1, 5))
    direct = (n - calc_bmin(n, 1, 50))
    cascades.append(cascade)
    directs.append(direct)

plt.figure(figsize(16, 9))
plt.plot(range(100), cascades)
plt.plot(range(100), directs)
```
