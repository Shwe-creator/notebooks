```python
%pylab inline
with open('test.txt', 'r') as f:
    lines = f.readlines()
```

```python
matsize = len(lines) // 10
data = np.zeros((9, matsize), dtype=np.int64)

for i in xrange(matsize):
    k = i * 10
    for j, l in enumerate(xrange(k + 1, k + 10)):
        data[j][i] = int(lines[l])
```

```python
plt.figure(figsize=(16, 5))
for i, label in zip(range(9), ['1/1', '2/7', '1/7', '1/3a', '1/3b', '3/1b', '3/1a', '7/1', '7/2']):
    data_at = data[i]
    print '{}: {}'.format(label, np.max(data_at))
    plt.plot(data_at, label=label)
plt.legend()
plt.xlim([0, 100])
```
